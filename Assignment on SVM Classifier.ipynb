## Encoding 

```python
x = df.drop('class',axis=1)
y = df['class']
```

```python
encoded_x = pd.get_dummies(x)
encoded_x
```

### train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(encoded_x,y,test_size=0.33, random_state=23)
```

```python
x_train
```

```python
y_train
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Model Building 

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'linear')
```

```python
svc.fit(x_train_scaled,y_train)
```

```python
y_pred = svc.predict(x_test_scaled)
```

## Performance Report

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

## HyperParameter tuning using GridSearchCV

```python
parameters = { 'kernel' : ['linear', 'poly', 'rbf'],
              'C' : [ 0.1, 1, 10],
              'gamma' : [ 1, 0.1, 0.01, 0.001]
             }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(SVC(), param_grid = parameters, verbose = 3, cv=2)
```

```python
clf.fit(x_train_scaled,y_train)
```

```python
clf.best_params_
```

```python
y_pred1 = clf.predict(x_test_scaled)
```

```python
print(classification_report(y_pred1,y_test))
```
