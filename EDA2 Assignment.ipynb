## LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()
```

```python
data['income'] = lab_enc.fit_transform(data['income'])
data['income']
```

 LabelEncoder assigns a numerical label to each category, while OneHotEncoder creates a binary vector representation of the categorical data, where each column represents a unique category.
 and the main difference is, in labelencoder rank is important and in onehot encoder rank is not important.

MinMaxScaler scales the data to a fixed range, typically between 0 and 1. On the other hand, StandardScaler rescales the data to have a mean of 0 and a standard deviation of 1.

## Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
clf = IsolationForest(random_state=12)
```

```python
clf.fit(enc)
```

```python
outlier = clf.predict(enc)
outlier
```

```python
enc['Score'] = clf.decision_function(enc)
```

```python
enc['Anomoly'] = clf.predict(enc.iloc[:,:-1])
```

```python
enc
```

```python
## outlier detection
enc[enc['Anomoly']==-1]
```
