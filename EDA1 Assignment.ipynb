```python
def finding_outlier(value):
    minimum,Q1,Q2,Q3,Maximum = np.quantile(value,[0,0.25,0.5,0.75,1.0])
    IQR = Q3-Q1
    lower_fence = Q1 - 1.5*(IQR)
    upper_fence = Q3 + 1.5*(IQR)

    outliers = []
    for i in value :
        if i >= lower_fence and i <= upper_fence:
            pass
        else:
            outliers.append(i)
    return outliers
```
