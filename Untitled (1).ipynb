```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Alphabets_data.csv')
df.head()
```

```python
df.info()
```

```python
## checking duplicated values
df[df.duplicated()]
```

```python
## deleting duplicated values
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
## split the data into dependent and independent feature 
x = df.drop('letter',axis=1)
y = df['letter']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
## split the data into train and test part
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=12)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Model Architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(x_train_scaled.shape[1],)),
    Dense(64, activation='relu'),
    Dense(len(set(y)), activation='softmax')
])
```

```python
model.summary()
```

```python
model.layers
```

```python
loss_function = 'sparse_categorical_crossentropy'
optimizer = 'SGD'
metrics = ['accuracy']

model.compile(optimizer= optimizer,
                  loss=loss_function,
                  metrics= metrics)
```

```python
## training
hist = model.fit(x_train_scaled,y_train, epochs=30,validation_split=0.2)
```

```python
new_d = pd.DataFrame(hist.history)
new_d
```

```python
new_d.plot()
```

```python
model.evaluate(x_test_scaled,y_test)
```

# HyperParameter tuning

```python
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop
import warnings
warnings.filterwarnings('ignore')
```

```python
# Function to create model, required for KerasClassifier
def create_model(optimizer='adam', init='glorot_uniform', neuron1=12, neuron2=8):
    model = Sequential()
    model.add(Dense(neuron1, input_shape=(16,), kernel_initializer=init, activation='relu'))
    model.add(Dense(neuron2, kernel_initializer=init, activation='relu'))
    model.add(Dense(3, kernel_initializer=init, activation='softmax'))
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
```

```python
# Create the KerasClassifier
model = KerasClassifier(model=create_model, verbose=0)
```

```python
# Define the grid search parameters
param_grid = {
    'model__optimizer': ['rmsprop', 'adam'],
    'model__init': ['glorot_uniform', 'normal', 'uniform'],
    'model__neuron1': [8, 12, 16],
    'model__neuron2': [4, 8, 12],
    'batch_size': [10, 20, 40],
    'epochs': [50, 100, 150]
}
```

```python
# Grid search
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(x_train_scaled, y_train)
```

```python
# Summarize results
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, std, param in zip(means, stds, params):
    print(f"{mean} ({std}) with: {param}")
```
